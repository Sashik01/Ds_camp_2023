```python
import numpy as np
import matplotlib.pyplot as plt


```

```python
def rand() -> list:
    return np.random.randint(1,10,20)
```

```python
def func(input_list: list) -> list:
    min_idx = np.argmin(input_list)
    a, b = input_list[:min_idx], input_list[min_idx:]
    max_idx1, max_idx2 = np.argmax(a), np.argmax(b)
    return np.array([max_idx1, max_idx2])
```

Розберись із кодом, ьо щось не так і візуалізацією
